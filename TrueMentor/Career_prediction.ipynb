```python
X=df.drop(['Career'],axis=1)
print(df.isnull().values.any())

X.head(20)
```

```python
y=df['Career'].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
X['What_is_your_educational_level?'].replace(to_replace ='post_matric',value = 0,inplace=True)
X['What_is_your_educational_level?'].replace(to_replace ='post_intermediate',value =1,inplace=True)
X['What_is_your_educational_level?'].replace(to_replace ='post_graduation',value =2,inplace=True)
X['What_is_your_educational_level?'].replace(to_replace ='professional',value =3,inplace=True)
X.replace(to_replace ='strongly dislike',value = 0,inplace=True)
X.replace(to_replace ='dislike',value = 1,inplace=True)
X.replace(to_replace ='unsure',value = 2,inplace=True)
X.replace(to_replace ='like',value = 3,inplace=True)
X.replace(to_replace ='strongly like',value = 4,inplace=True)

X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
```

```python
regressor=DecisionTreeClassifier()
cv = KFold(n_splits=10, random_state=42, shuffle=False)

regressor.fit(X_train,y_train)
```

```python
#print(X_test.iloc[0])
#print(y_test.iloc[0])
y_pred = regressor.predict(X_test)
y_pred
#regressor.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=51)
rf_classifier.fit(X_train,y_train)
y_pred_rf=rf_classifier.predict(X_test)
accuracy_score(y_test,y_pred_rf)
```
